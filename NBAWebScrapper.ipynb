### Team Roster

```python
def get_roster(team_id):
    
    header = {
        "Accept": "application/json, text/plain, */*",
        "Accept-Encoding": "gzip, deflate, br",
        "Accept-Language": "en-GB,en-US;q=0.9,en;q=0.8",
        "Cache-Control": "no-cache",
        "Connection": "keep-alive",
        "Host": "stats.nba.com",
        "Pragma": "no-cache",
        "Referer": "https://stats.nba.com/team/{}/".format(team_id),
        "Sec-Fetch-Dest": "empty",
        "Sec-Fetch-Mode": "cors",
        "Sec-Fetch-Site": "same-origin",
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.105 Safari/537.36",
        "x-nba-stats-origin": "stats",
        "x-nba-stats-token": "true",
        "X-NewRelic-ID": "VQECWF5UChAHUlNTBwgBVw=="
            }

    params = (
        ("LeagueID", "00"),
        ("Season", "2019-20"),
        ("TeamID", "{}".format(team_id)),
        ("resource","commonteamroster") )
    
    response = requests.get('https://stats.nba.com/stats/commonteamroster', headers=header, params=params)

    jsn = response.json()
    frame = pd.DataFrame(jsn['resultSets'][0]['rowSet'])
    frame.columns = jsn['resultSets'][0]['headers']
    return frame
```

```python
roster_byteam = {}
for i in team_ids:
    roster = get_roster(i)
    team_name = df[df.teamId == "{}".format(i)].teamShortName.iloc[0]
    print("Roster scrapped succesfully for {}".format(team_name))
    roster_byteam[team_name] = roster
    time.sleep(1)
```

```python
filename = 'roster'
outfile = open(filename,'wb')
pickle.dump(roster_byteam,outfile)
outfile.close()
```
